```python
import requests

url = "https://raw.githubusercontent.com/disposable-email-domains/disposable-email-domains/refs/heads/main/disposable_email_blocklist.conf"
resp = requests.get(url)
resp.raise_for_status()

raw_text = resp.text
lines = raw_text.strip().split("\n")

print(f"Total lines: {len(lines)}")
print("First 10 lines:")
for line in lines[:10]:
    print(repr(line))
```

```python
rows = [{"domain": line.strip()} for line in lines if line.strip() != ""]

domains_df = spark.createDataFrame(rows)
display(domains_df.limit(10))
print(f"Total rows: {domains_df.count()}")

domains_df.write.mode("overwrite").saveAsTable("fraud_detection.bronze.disposable_domains_raw")

display(spark.sql("SELECT COUNT(*) FROM fraud_detection.bronze.disposable_domains_raw"))
```
