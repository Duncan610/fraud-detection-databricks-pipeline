```python
import requests

url = "https://raw.githubusercontent.com/stamparm/ipsum/master/ipsum.txt"
resp = requests.get(url)
resp.raise_for_status()

raw_text = resp.text
lines = raw_text.strip().split("\n")

print(f"Total lines: {len(lines)}")
print("First 10 lines:")
for line in lines[:10]:
    print(repr(line))
```

```python
data_lines = [line for line in lines if not line.startswith("#") and line.strip() != ""]

print(f"Data lines after filtering comments: {len(data_lines)}")
print(data_lines[:5])
```

```python
rows = []
for line in data_lines:
    parts = line.split("\t")
    rows.append({
        "ip_address": parts[0],
        "blacklist_count": int(parts[1])
    })

blocklist_df = spark.createDataFrame(rows)
display(blocklist_df.limit(10))
print(f"Total rows: {blocklist_df.count()}")
```

```python
blocklist_df.write.mode("overwrite").saveAsTable("fraud_detection.bronze.ip_blocklist_raw")
display(spark.sql("SELECT count(*) FROM fraud_detection.bronze.ip_blocklist_raw"))
```
